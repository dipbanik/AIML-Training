```
import numpy as np
class Perceptron(object):
    def __init__(self, no_of_cols=1, no_of_rows=100, learning_rate=0.01,\
                 epochs=100):
        self.no_of_rows = no_of_rows
        self.learning_rate = learning_rate
        self.weights = np.zeros(no_of_cols + 1)
        self.epochs = epochs
    
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation >=0:
            activation =1 
        else:
            activation = 0
        return activation
    
    def train(self, training_inputs, labels):
        for i in range(self.epochs):
            print("Epoch :", i+1)
            print('Weights :', self.weights)
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs) # calling activation on input
                #update weights
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                #update bias
                self.weights[0] += self.learning_rate * (label - prediction)
```

```
import pandas as pd
df = pd.read_csv('iris.csv')
```

```
df.columns
```

```
X=df[['sepal_len', 'sepal_wid', 'petal_len']]
X=X.values
Y=df['species']
```

```
Y= Y.map(lambda x: 0 if x== "Iris-setosa" else 1)
```

```
no_of_cols=X.shape[1] #3
no_of_rows=len(X) #150
learning_rate=0.01
epochs=18
clf = Perceptron(no_of_cols, no_of_rows, learning_rate, epochs)
```

```
clf.train(X,Y)
```

```
clf.weights
```

```
clf.predict(X[0])
```

```
Y[0]
```

```
pred = []
for i in range(0,150):
  pred.append(clf.predict(X[i]))
```

```
for real,p in zip(Y,pred):
  print("Real", real,"predicted",p)
  
```

```
from sklearn.metrics import accuracy_score 
```

```
accuracy_score(Y,pred)
```
